```python
import hashlib
import os
from ast import Try
ALGORITMI = ['md5', 'sha1', 'sha1', 'sha256', 'sha512']
```

```python
def calcola_hash(testo, algoritmo='md5'):
  if algoritmo not in ALGORITMI:
    print("Algoritmo non supportato. scegli tra: {ALGORITMI}")
    return None
  h = hashlib.new(algoritmo)
  h.update(testo.encode('utf-8'))
  return h.hexdigest()

def identifica_algoritmo(hash_string):
  lunghezza = len(hash_string)
  mappa = {
      32: 'MD5',
      40: 'SHA1',
      64: 'SHA256',
      128: 'SHA512'
  }

  return mappa.get(lunghezza, 'Algoritmo sconosciuto')

parole = ["ciao", "password", "admin", "123456"]

print(f"{'Parola':<12} {'MD5':<35} {'SHA256':<20}")
print("-"* 90)
for parola in parole:
  md5 = calcola_hash(parola, 'md5')
  sha256 = calcola_hash(parola, 'sha256')[:20] + "..."
  print(f"{parola:<12} {md5:<35} {sha256:<20}")
```

```python
def genera_varianti(parola):
  varianti = set()
  varianti.add(parola)
  varianti.add(parola.lower())
  varianti.add(parola.upper())
  varianti.add(parola.capitalize())

  leet = parola.lower()
  leet = leet.replace('a', '4').replace('e', '3')
  leet = leet.replace('i', '1').replace('o', '0').replace('s', '5')
  varianti.add(leet)
  varianti.add(leet.capitalize())

  for numero in ['1', '12', '123', '1234', '!', '123!', '1!', '12!']:
    varianti.add(parola + numero)
    varianti.add(parola.capitalize() + numero)


  for anno in ['1926', '1990', '1995', '2000', '2001', '2010', '2024', '2025']:
    varianti.add(parola + anno)
    varianti.add(parola.capitalize() + anno)
    varianti.add(parola + anno + "!")
    varianti.add(parola.capitalize() + anno + "!")


  return list(varianti)

def attacco_con_regole(hash_target, wordlist_base, algoritmo= 'md5'):
  print("="*55)
  print(f" ATTACCO CON REGOLE (hashcat-style)")
  print(f" Parole base : {len(wordlist_base)}")

  wordlist_espansa = []
  for parola in wordlist_base:
    wordlist_espansa.extend(genera_varianti(parola.strip()))


  wordlist_espansa = list(set(wordlist_espansa))
  print(f" Dopo regole : {len(wordlist_espansa)} varianti")
  print("="*55)

  for i, parola in enumerate(wordlist_espansa):
    hash_tentativo = calcola_hash(parola, algoritmo)
    if hash_tentativo == hash_target:
      print(f"\n TROVATA! Password: '{parola}'")
      print(f" Tentativi: {i+1}/{len(wordlist_espansa)}")
      print("="*55)
      return parola
  print(f"\n Non trovata con {len(wordlist_espansa)} varianti.")
  print("="*55)
  return None

hash_con_regole = [
    (calcola_hash("Ciao123", "md5"),    "md5",    "Ciao123"),
    (calcola_hash("pizza2024", "sha256"), "sha256", "pizza2024"),
    (calcola_hash("L4z10", "md5"),      "md5",    "L4z10"),
]

parole_base = ["ciao", "pizza", "lazio", "calcio", "admin"]

for hash_t, algo, risposta in hash_con_regole:
  print(f"\nRisposta attesa: '{risposta}")
  attacco_con_regole(hash_t, parole_base, algo)
```

```python
import hashlib

def calcola_hash(testo, algoritmo='md5'):
    h = hashlib.new(algoritmo)
    h.update(testo.encode('utf-8'))
    return h.hexdigest()

def genera_varianti_v2(parola):
    varianti = set()
    varianti.add(parola)
    varianti.add(parola.lower())
    varianti.add(parola.upper())
    varianti.add(parola.capitalize())
    leet = parola.lower()
    leet = leet.replace('a','4').replace('e','3')
    leet = leet.replace('i','1').replace('o','0').replace('s','5')
    varianti.add(leet)
    varianti.add(leet.capitalize())
    for numero in ['1', '12', '123', '1234', '!', '123!']:
        varianti.add(parola + numero)
        varianti.add(parola.capitalize() + numero)
    for anno in ['1926', '1990', '1995', '2000', '2010', '2024', '2025']:
        varianti.add(parola + anno)
        varianti.add(parola.capitalize() + anno)
        varianti.add(parola + anno + "!")
        varianti.add(parola.capitalize() + anno + "!")
    return list(varianti)

def attacco_finale(hash_target, wordlist_base, algoritmo='md5'):
    print("=" * 55)
    print(f"  ATTACCO FINALE")
    print(f"  Parole base   : {len(wordlist_base)}")
    wordlist_espansa = []
    for parola in wordlist_base:
        wordlist_espansa.extend(genera_varianti_v2(parola.strip()))
    wordlist_espansa = list(set(wordlist_espansa))
    print(f"  Dopo regole   : {len(wordlist_espansa)} varianti")
    print("=" * 55)
    for i, parola in enumerate(wordlist_espansa):
        if calcola_hash(parola, algoritmo) == hash_target:
            print(f"\n  TROVATA! Password: '{parola}'")
            print(f"  Tentativi: {i+1}/{len(wordlist_espansa)}")
            print("=" * 55)
            return parola
    print(f"\n  Non trovata con {len(wordlist_espansa)} varianti.")
    print("=" * 55)
    return None

# wordlist
parole_base_sfida = [
    "ciao", "pizza", "calcio", "napoli", "milan", "roma",
    "lazio", "inter", "juventus", "italia", "admin", "password",
    "sole", "mare", "amore", "casa", "mario", "luigi",
    "secret", "123456", "letmein", "dragon", "master", "shadow",
    "monkey", "sunshine", "football", "qwerty", "welcome",
    "hello", "ninja", "wizard", "batman", "superman"
]

# hash generati direttamente qui
parola1 = "hello"
parola2 = "123456"
parola3 = "Napoli1926!"

sfide_corrette = [
    (calcola_hash(parola1, "md5"),    "md5",    parola1),
    (calcola_hash(parola2, "md5"),    "md5",    parola2),
    (calcola_hash(parola3, "sha256"), "sha256", parola3),
]

print("Hash generati:\n")
for hash_t, algo, parola in sfide_corrette:
    print(f"  {parola:<15} → {hash_t}")

print("\nSFIDA FINALE\n")
for i, (hash_t, algo, _) in enumerate(sfide_corrette):
    print(f"Hash {i+1}:")
    attacco_finale(hash_t, parole_base_sfida, algo)
    print()
```
