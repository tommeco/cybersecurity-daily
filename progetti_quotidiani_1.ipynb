```python
"STEP 1 - FUNZIONE DI CIFRATURA"
def cifra(testo, shift):
  risultato = []
  shift = shift % 26

  for carattere in testo:
    if carattere.isalpha():
      base = ord('A') if carattere.isupper() else ord ('a')
      nuovo_char = chr((ord(carattere)- base + shift)% 26 + base)
      risultato.append(nuovo_char)
    else:
      risultato.append(carattere)
  return "".join(risultato)
```

```python
print(cifra("Ciao Mondo!", 3 ))
```

```python
"STEP 2 - FUNZIONE DI DECIFRATURA"
# 1. Devi prima definire come si cifra!
def cifra(testo, shift):
    risultato = ""
    for carattere in testo:
        if carattere.isalpha():
            # Gestisce lo shift restando nell'alfabeto
            start = ord('a') if carattere.islower() else ord('A')
            nuovo_carattere = chr((ord(carattere) - start + shift) % 26 + start)
            risultato += nuovo_carattere
        else:
            risultato += carattere
    return risultato

# 2. Poi definisci decifra (che usa la funzione sopra)
def decifra(testo_cifrato, shift):
    return cifra(testo_cifrato, -shift)

# 3. Test del codice
messaggio_originale = "Cybersecurity è fantastica"
chiave = 13

# Esecuzione passaggi
testo_segreto = cifra(messaggio_originale, chiave)
testo_chiaro = decifra(testo_segreto, chiave)

print(f"Cifrato: {testo_segreto}")
print(f"Decifrato: {testo_chiaro}")
```

```python
'STEP 4 - BRUTE FORCE INTELLIGENTE CON ANALISI DELLE FREQUENZE'
from collections import Counter
LETTERE_FREQUENTI_IT = 'eaoirstnlcupdmgvhfbzqkjwxy'
LETTERE_FREQUENTI_EN = 'etaoinshrdlcumwfgypbvkjxqz'

def indovina_shift(testo_cifrato, lingua='it'):
  ref =LETTERE_FREQUENTI_IT if lingua == 'it' else LETTERE_FREQUENTI_EN
  sole_lettere = [c.lower() for c in testo_cifrato if c.isalpha()]
  if not sole_lettere:
    return None

  conteggio = Counter(sole_lettere)
  lettera_piu_frequente = conteggio.most_common(1)[0][0]

  shift_stimato = (ord(lettera_piu_frequente)- ord(ref[0])) % 26

  print(f"Lettera più frequente nel cifrato: '{lettera_piu_frequente}'")
  print(f"lettera più frequente attesa ({lingua}):'{ref[0]}'")
  print(f"shift stimato: {shift_stimato}")

  candidato = decifra(testo_cifrato, shift_stimato)
  print(f"Testo decifrato: {candidato}")
  return shift_stimato

testo_lungo =(
      "Ocz nzxgzo dn: xvrznajxf dn m jvo voo vxf!")
chiave_segreta = 7

cifrato_lungo = cifra(testo_lungo, chiave_segreta)

print(f"Testo cifrato (shift={chiave_segreta}):\n{cifrato_lungo}\n")
print("-"* 55)

shift_trovato = indovina_shift(cifrato_lungo, lingua='it')
print(f"\nChiave reale: {chiave_segreta} | Trovata: {shift_trovato} | OK : {chiave_segreta == shift_trovato}")
```
